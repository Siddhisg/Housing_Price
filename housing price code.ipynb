Ridge Regression From Scratch 

```python
# Implementing Ridge Regression from Scratch
class RidgeRegressionScratch:
    def __init__(self, alpha=1.0):
        self.alpha = alpha  # Regularization strength
        self.weights = None

    def fit(self, X, y):
        # Add bias term to the input data
        X_bias = np.c_[np.ones((X.shape[0], 1)), X]
        # Calculate weights using the regularized normal equation: (X^T * X + alpha * I)^(-1) * X^T * y
        X_transpose = X_bias.T
        identity_matrix = np.eye(X_bias.shape[1])
        self.weights = np.linalg.inv(X_transpose @ X_bias + self.alpha * identity_matrix) @ X_transpose @ y

    def predict(self, X):
        # Add bias term to the input data
        X_bias = np.c_[np.ones((X.shape[0], 1)), X]
        return X_bias @ self.weights

    def evaluate(self, X, y):
        predictions = self.predict(X)
        mse = np.mean((y - predictions) ** 2)
        rmse = np.sqrt(mse)
        r2 = 1 - (np.sum((y - predictions) ** 2) / np.sum((y - np.mean(y)) ** 2))
        return mse, rmse, r2

# Evaluate the Ridge Regression model on a given dataset
def evaluate_ridge_model_on_dataset(X_train, X_test, y_train, y_test, alpha=1.0):
    model = RidgeRegressionScratch(alpha=alpha)
    model.fit(X_train, y_train)
    mse, rmse, r2 = model.evaluate(X_test, y_test)
    return mse, rmse, r2

# Boston Housing Dataset with Ridge Regression
boston_ridge_mse, boston_ridge_rmse, boston_ridge_r2 = evaluate_ridge_model_on_dataset(
    X_train_scaled, X_test_scaled, y_train.values, y_test.values, alpha=1.0)

# Ames Housing Dataset with Ridge Regression
ames_ridge_mse, ames_ridge_rmse, ames_ridge_r2 = evaluate_ridge_model_on_dataset(
    X_train_ames_scaled_df.values, X_test_ames_scaled_df.values, y_train_ames.values, y_test_ames.values, alpha=1.0)

# Prepare the results for all datasets
ridge_results = {
    'Boston Housing (Ridge Regression)': {'MSE': boston_ridge_mse, 'RMSE': boston_ridge_rmse, 'R2': boston_ridge_r2},
    # 'California Housing (Ridge Regression)': {'MSE': california_ridge_mse, 'RMSE': california_ridge_rmse, 'R2': california_ridge_r2},
    'Ames Housing (Ridge Regression)': {'MSE': ames_ridge_mse, 'RMSE': ames_ridge_rmse, 'R2': ames_ridge_r2},
}

print("Ridge Regression Model Evaluation:")
for dataset, metrics in ridge_results.items():
    print(f"\n{dataset}:")
    for metric, value in metrics.items():
        print(f"  {metric}: {value}")
```

Cross-Validation on Ridge Regression 

```python
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
import numpy as np

def custom_cross_val_score(model_class, X, y, alpha, degree=1, cv=5):
    kf = KFold(n_splits=cv, shuffle=True, random_state=42)
    mse_scores = []
    rmse_scores = []
    r2_scores = []
    
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        # Apply polynomial features with lower degree
        poly = PolynomialFeatures(degree=degree, include_bias=False)
        X_train_poly = poly.fit_transform(X_train)
        X_test_poly = poly.transform(X_test)
        
        # Apply standard scaling
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train_poly)
        X_test_scaled = scaler.transform(X_test_poly)
        
        # Instantiate and train the model with higher regularization
        model = model_class(alpha=alpha)
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        
        # Calculate metrics
        mse = np.mean((y_test - y_pred) ** 2)
        rmse = np.sqrt(mse)
        r2 = 1 - (np.sum((y_test - y_pred) ** 2) / np.sum((y_test - np.mean(y_test)) ** 2))
        
        mse_scores.append(mse)
        rmse_scores.append(rmse)
        r2_scores.append(r2)
    
    return np.mean(mse_scores), np.mean(rmse_scores), np.mean(r2_scores)

# Perform cross-validation on Boston Housing dataset with lower degree polynomial and increased alpha
best_alpha_boston = 10.0 # Increased regularization
X_train_boston_simpler = X_train_scaled_df.values
mse_boston_cv, rmse_boston_cv, r2_boston_cv = custom_cross_val_score(
    RidgeRegressionScratchImproved, X_train_boston_simpler, y_train.values, alpha=best_alpha_boston, degree=1, cv=2)

# Perform cross-validation on Ames Housing dataset with lower degree polynomial and increased alpha
best_alpha_ames = 10.0  # Increased regularization
X_train_ames_simpler = X_train_ames_selected.values
mse_ames_cv, rmse_ames_cv, r2_ames_cv = custom_cross_val_score(
    RidgeRegressionScratchImproved, X_train_ames_simpler, y_train_ames.values, alpha=best_alpha_ames, degree=1, cv=5
)

# Output the cross-validation results
cross_val_results = {
    'Boston Housing (Cross-Validation)': {'MSE': mse_boston_cv, 'RMSE': rmse_boston_cv, 'R2': r2_boston_cv},
    'Ames Housing (Cross-Validation)': {'MSE': mse_ames_cv, 'RMSE': rmse_ames_cv, 'R2': r2_ames_cv},
}

print("Cross-Validation Results:")
for dataset, metrics in cross_val_results.items():
    print(f"\n{dataset}:")
    for metric, value in metrics.items():
        print(f"  {metric}: {value}")
```

Decision Tree Regressor Model From Scratch

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

# Decision Tree Regressor from scratch
class DecisionTreeRegressorScratch:
    def __init__(self, max_depth=5, min_samples_split=2):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.tree = None

    def fit(self, X, y):
        self.tree = self._build_tree(X, y, depth=0)

    def _build_tree(self, X, y, depth):
        num_samples, num_features = X.shape
        if depth >= self.max_depth or num_samples < self.min_samples_split:
            return np.mean(y)
        
        best_mse = float('inf')
        best_split = None
        best_splits = None

        for feature_index in range(num_features):
            thresholds = np.unique(X[:, feature_index])
            for threshold in thresholds:
                left_mask = X[:, feature_index] <= threshold
                right_mask = ~left_mask
                
                if np.sum(left_mask) == 0 or np.sum(right_mask) == 0:
                    continue
                
                left_y, right_y = y[left_mask], y[right_mask]
                mse = self._calculate_mse(left_y, right_y)
                
                if mse < best_mse:
                    best_mse = mse
                    best_split = (feature_index, threshold)
                    best_splits = (left_mask, right_mask)
        
        if best_split is None:
            return np.mean(y)
        
        left_tree = self._build_tree(X[best_splits[0]], y[best_splits[0]], depth + 1)
        right_tree = self._build_tree(X[best_splits[1]], y[best_splits[1]], depth + 1)
        
        return {'split_feature': best_split[0], 'split_value': best_split[1], 
                'left': left_tree, 'right': right_tree}

    def _calculate_mse(self, left_y, right_y):
        left_mse = np.var(left_y) * len(left_y)
        right_mse = np.var(right_y) * len(right_y)
        return (left_mse + right_mse) / (len(left_y) + len(right_y))

    def predict(self, X):
        return np.array([self._predict_single(x, self.tree) for x in X])

    def _predict_single(self, x, tree):
        if not isinstance(tree, dict):
            return tree
        
        split_feature = tree['split_feature']
        split_value = tree['split_value']
        
        if x[split_feature] <= split_value:
            return self._predict_single(x, tree['left'])
        else:
            return self._predict_single(x, tree['right'])

# Instantiate and train the model
dt_regressor = DecisionTreeRegressorScratch(max_depth=5, min_samples_split=10)
dt_regressor.fit(X_train_scaled, y_train)

# Predictions
y_pred_scratch = dt_regressor.predict(X_test_scaled)

# Calculate evaluation metrics
mse_scratch = mean_squared_error(y_test, y_pred_scratch)
rmse_scratch = np.sqrt(mse_scratch)
r2_scratch = r2_score(y_test, y_pred_scratch)

# Train the Decision Tree Regressor from scratch on Ames Housing dataset
dt_regressor_ames = DecisionTreeRegressorScratch(max_depth=5, min_samples_split=10)
dt_regressor_ames.fit(X_train_ames_scaled_df.values, y_train_ames.values)

# Predictions
y_pred_scratch_ames = dt_regressor_ames.predict(X_test_ames_scaled_df.values)

# Calculate evaluation metrics
mse_scratch_ames = mean_squared_error(y_test_ames, y_pred_scratch_ames)
rmse_scratch_ames = np.sqrt(mse_scratch_ames)
r2_scratch_ames = r2_score(y_test_ames, y_pred_scratch_ames)

# Print the evaluation metrics
print("Ames Housing Dataset:")
print(f"MSE: {mse_scratch_ames}")
print(f"RMSE: {rmse_scratch_ames}")
print(f"R^2: {r2_scratch_ames}")

# Print the evaluation metrics
print("Boston Housing Dataset:")
print(f"MSE: {mse_scratch}")
print(f"RMSE: {rmse_scratch}")
print(f"R^2: {r2_scratch}\n")
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

# Random Forest Regressor from scratch
class RandomForestRegressorScratch:
    def __init__(self, n_trees=10, max_depth=5, min_samples_split=10, max_features=None):
        self.n_trees = n_trees
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.max_features = max_features
        self.trees = []

    def _bootstrap_sample(self, X, y):
        n_samples = X.shape[0]
        indices = np.random.choice(n_samples, size=n_samples, replace=True)
        return X[indices], y[indices]

    def _feature_subsample(self, X):
        n_features = X.shape[1]
        if self.max_features is None:
            max_features = n_features
        else:
            max_features = min(self.max_features, n_features)
        
        feature_indices = np.random.choice(n_features, size=max_features, replace=False)
        return feature_indices

    def fit(self, X, y):
        self.trees = []
        for _ in range(self.n_trees):
            X_sample, y_sample = self._bootstrap_sample(X, y)
            feature_indices = self._feature_subsample(X_sample)
            X_sample = X_sample[:, feature_indices]
            
            tree = DecisionTreeRegressorScratch(max_depth=self.max_depth, min_samples_split=self.min_samples_split)
            tree.fit(X_sample, y_sample)
            self.trees.append((tree, feature_indices))

    def predict(self, X):
        tree_predictions = np.array([
            tree.predict(X[:, feature_indices]) for tree, feature_indices in self.trees
        ])
        return np.mean(tree_predictions, axis=0)

# boston
# Load and prepare the data (assuming the data is already loaded and preprocessed)
X_train_boston_np = X_train_scaled_df.values
y_train_boston_np = y_train.values
X_test_boston_np = X_test_scaled_df.values
y_test_boston_np = y_test.values

# Instantiate and train the Random Forest model
rf_regressor = RandomForestRegressorScratch(n_trees=30, max_depth=5, min_samples_split=5, max_features=5)
rf_regressor.fit(X_train_boston_np, y_train_boston_np)

# Predictions
y_pred_rf_scratch = rf_regressor.predict(X_test_boston_np)

# Calculate evaluation metrics
mse_rf_scratch = mean_squared_error(y_test_boston_np, y_pred_rf_scratch)
rmse_rf_scratch = np.sqrt(mse_rf_scratch)
r2_rf_scratch = r2_score(y_test_boston_np, y_pred_rf_scratch)

# Print metrics for Boston Housing dataset
print(f"Boston Housing Dataset:")
print(f"  Mean Squared Error (MSE) :{mse_rf_scratch:.4f}")
print(f"  Root Mean Squared Error (RMSE) :{rmse_rf_scratch:.4f}")
print(f"  R-squared (R²) : {r2_rf_scratch:.4f}\n")

#ames
# Instantiate and train the Random Forest model for Ames Housing dataset
X_train_ames_np = X_train_ames_scaled_df.values
y_train_ames_np = y_train_ames.values
X_test_ames_np = X_test_ames_scaled_df.values
y_test_ames_np = y_test_ames.values

rf_regressor_ames = RandomForestRegressorScratch(n_trees=100, max_depth=10, min_samples_split=3, max_features=5)
rf_regressor_ames.fit(X_train_ames_np, y_train_ames_np)

# Predictions
y_pred_rf_scratch_ames = rf_regressor_ames.predict(X_test_ames_np)

# Calculate evaluation metrics
mse_rf_scratch_ames = mean_squared_error(y_test_ames_np, y_pred_rf_scratch_ames)
rmse_rf_scratch_ames = np.sqrt(mse_rf_scratch_ames)
r2_rf_scratch_ames = r2_score(y_test_ames_np, y_pred_rf_scratch_ames)

# Print metrics for Ames Housing dataset
print(f"Ames Housing Dataset:")
print(f"  Mean Squared Error (MSE) :{mse_rf_scratch_ames:.4f}")
print(f"  Root Mean Squared Error (RMSE) : {rmse_rf_scratch_ames:.4f}")
print(f"  R-squared (R²) :{r2_rf_scratch_ames:.4f}")
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

class BaggingRegressorScratch:
    def __init__(self, base_model, n_estimators=10, max_depth=5, min_samples_split=10, max_features=None):
        self.base_model = base_model
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.max_features = max_features
        self.models = []

    def _bootstrap_sample(self, X, y):
        n_samples = X.shape[0]
        indices = np.random.choice(n_samples, size=n_samples, replace=True)
        return X[indices], y[indices]

    def fit(self, X, y):
        self.models = []
        for _ in range(self.n_estimators):
            X_sample, y_sample = self._bootstrap_sample(X, y)
            model = self.base_model(max_depth=self.max_depth, min_samples_split=self.min_samples_split)
            model.fit(X_sample, y_sample)
            self.models.append(model)

    def predict(self, X):
        predictions = np.array([model.predict(X) for model in self.models])
        return np.mean(predictions, axis=0)

# Apply Bagging on the datasets
def apply_bagging(X_train, y_train, X_test, y_test, dataset_name):
    bagging_regressor = BaggingRegressorScratch(
        base_model=DecisionTreeRegressorScratch,
        n_estimators=50, max_depth=10, min_samples_split=5
    )
    bagging_regressor.fit(X_train, y_train)
    y_pred_bagging = bagging_regressor.predict(X_test)

    mse_bagging = mean_squared_error(y_test, y_pred_bagging)
    rmse_bagging = np.sqrt(mse_bagging)
    r2_bagging = r2_score(y_test, y_pred_bagging)

    print(f"{dataset_name} Dataset with Bagging:")
    print(f"  Mean Squared Error (MSE): {mse_bagging:.4f}")
    print(f"  Root Mean Squared Error (RMSE): {rmse_bagging:.4f}")
    print(f"  R-squared (R²): {r2_bagging:.4f}\n")

# Bagging on Boston Housing Dataset
apply_bagging(X_train_boston_np, y_train_boston_np, X_test_boston_np, y_test_boston_np, "Boston Housing")

# Bagging on Ames Housing Dataset
apply_bagging(X_train_ames_np, y_train_ames_np, X_test_ames_np, y_test_ames_np, "Ames Housing")
```

Support Vector Regressor (SVR) 

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np

# Function to evaluate SVR with different kernels
def evaluate_svr_kernels(X_train, X_test, y_train, y_test):
    kernels = ['linear', 'poly', 'rbf']
    results = []
    
    for kernel in kernels:
        # Initialize and train SVR model
        svr = SVR(kernel=kernel)
        svr.fit(X_train, y_train)
        
        # Predictions
        y_pred = svr.predict(X_test)
        
        # Calculate evaluation metrics
        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_test, y_pred)
        
        # Store results
        results.append({
            'Kernel': kernel,
            'MSE': mse,
            'RMSE': rmse,
            'R²': r2
        })
    
    return results

# Example function to print results
def print_svr_results(results, dataset_name):
    print(f"\n{dataset_name} SVR Kernel Comparison:")
    for result in results:
        print(f"Kernel: {result['Kernel']}")
        print(f"  MSE: {result['MSE']:.4f}")
        print(f"  RMSE: {result['RMSE']:.4f}")
        print(f"  R²: {result['R²']:.4f}")
        print("-" * 30)

# Boston Housing Dataset
results_boston = evaluate_svr_kernels(X_train_scaled_df, X_test_scaled_df, y_train, y_test)
print_svr_results(results_boston, "Boston Housing")

# Ames Housing Dataset
results_ames = evaluate_svr_kernels(X_train_ames_scaled_df, X_test_ames_scaled_df, y_train_ames, y_test_ames)
print_svr_results(results_ames, "Ames Housing")
```
